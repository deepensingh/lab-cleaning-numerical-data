# Rename the columns so they follow the PE8 (snake case) (lower_case_with_underscores)

```python
cols = [] # Initialize an empty list cols to store the modified column names.
for i in range(len(customer_df.columns)): # Loop over each column name in "customer_df" using "range function (len(customer_df.columns))"
    cols.append(customer_df.columns[i].lower().replace(' ', '_')) # methods --> .lower() and .replace(' ', '_') on each column
customer_df.columns = cols
```

```python
customer_df.head(10)
```

# Change effective to date column to datetime format.

# Explanation:

The pd.to_datetime() function is used to convert a column of a DataFrame (or a Series) to a datetime format.
The input of the function is the column that you want to convert. In this case, it's the "effective_to_date" column of the 

DataFrame.
The output of the function is the same column in a datetime format.
The last line of code, df['effective_to_date'] = pd.to_datetime(df['effective_to_date']) replace the old column with the new converted column.

```python
# effective_to_date it is an object so we will change into the date format
```

# Begin with Numerical Data, 
Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
# Checking the unique values in the numerical columns
for col in numerical.columns:
    print(len(customer_df[col].unique()))
display(customer_df.shape[0] * 0.02)
```

```python
# Dividing into discreate list and continuous list
discrete_lst = []
continuous_lst = []
for col in numerical.columns:
    if len(customer_df[col].unique()) < (customer_df.shape[0] * 0.02):
        discrete_lst.append(col)
    else:
        continuous_lst.append(col)
```

```python
print(discrete_lst)
print(continuous_lst)
```

# Plot a correlation matrix, comment what you see.

```python
correlation = customer_df.corr() 
correlation
```

```python
fig, ax=plt.subplots(figsize=(10,8))
ax=sns.heatmap(correlation, annot=True)
plt.show()
```

All the values have low correlation, except for the pair of "total_claim_amount" and "monthly_premium_auto" with a correlation of 0.63, and "monthly_premium_auto" with "customer_lifetime_value." If we must remove a column, I suggest "monthly_premium_auto" as it has the highest correlation with two variables.

```python
for column in numerical[discrete_lst]:
    sns.distplot(numerical[column])
    plt.show()
```

```python
for column in numerical[continuous_lst]:
    sns.distplot(numerical[column])
    plt.show()
```

# Comment what you can see in the plots.

Outliers can be seen in the discrete variable "monthly_premium_auto" and in the continuous variables "customer_lifetime_value," "income," and "total_claim_amount."
